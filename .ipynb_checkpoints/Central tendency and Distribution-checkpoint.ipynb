# STATISTICAL PRACTICAL IMPLEMENTATION
Measure of central Tendency
1) Mean
2) Mode
3) Median

```python
ages = [23,24,32,45,12,43,67,45,32,56,32,120]
```

```python
import numpy as np
```

```python
mean = np.mean(ages)
median = np.median(ages)
print("The mean is {} and the median is {} ".format(mean,median))
# Outlier does not create any impact on median so here median remain same where as mean changes to 44.25
```

```python
import statistics 
print(statistics.mean(ages))
print(statistics.median(ages))
print(statistics.mode(ages))     # Mode is used where we found more categorical variable(like female, specific flower etc).
```

```python
import seaborn as sns
sns.boxplot(ages,orient='h')       # five number summary- which include(minimum , Q1 , median, Q3, Maximum)
```

# Five number summary

```python
q1,q3 = np.percentile(ages,[25,75])
print(q1,q3)
```

```python
iqr = q3-q1
print(iqr)
```

```python
lf = q1 - (1.5*iqr)
hf = q3 + (1.5*iqr)
print(lf)
print(hf)
```

# Mesure of dispersion
1) Variance
2) Standard deviation

```python
statistics.variance(ages)   # its uses the formula of sample variance
```

```python
np.var(ages,axis=0)   # This is the population variance
```

```python
def variance(data):
    n=len(ages)
    mean = sum(data)/n    # population variance
    ## variance
    deviation = [(x- mean)**2 for x in data]
    variance = sum(deviation)/n
    return variance    
```

```python
variance(ages)
```

```python
def variance(data,dof):
    n=len(ages)
    mean = sum(data)/n    # population variance
    ## variance
    deviation = [(x- mean)**2 for x in data]
    variance = sum(deviation)/n-dof
    return variance    
```

```python
variance(ages,1)
```

```python
statistics.pvariance(ages)   # population variance
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

# Created log Normal distribution

```python
mu,sigma = 3,1   # mu = mean , sigma = std deviation
log_normal = np.random.lognormal(mu,sigma,1000)
print(log_normal)
```

```python
sns.histplot(log_normal,kde=True)    # This is log normal distribution
```

```python
# applying log in log normal distribution to convert it to normal distrubtion
sns.histplot(np.log(log_normal),kde=True)     # Convert log normal distribution to Normal distribution
```

# Check Whether the distrubtion is normal ditribution or not 

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import seaborn as sns
import pylab
```

```python
# if you want to check whether the feature is gaussian or normal distrubuted.
# Q-Q plot
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist="norm",plot=pylab)
    plt.show()
```

```python
s= np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
mu,sigma = 3,1
sample = np.random.lognormal(mu,sigma,100)
```

```python
def plot_data_log(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(np.log(sample),dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data_log(sample)
```
